** Задача 1**
Покажите, что ROC-AUC  в случае, когда классификатор даёт случайные ответы $a(x) = 1$ с вероятностью $p$ и $a(x) = 0$ с вероятностью $1 − p$, будет в среднем равен $0.5$ независимо от $p$ и доли класса $1$ в обучающей выборке.

**Решение:** В данном случае линейный классификатор можно представить как $a(x) = \text{threshold}(l(x))$, где $l(x) \in [0; 1]$, $\text{threshold}_p$ -- пороговая функция с параметром $p$. Варьируя параметр $p$, начнем строить ROC-кривую. Заметим, что в таком случае параметр $p$ пробегает значения от $0$ до $1$, по сути это означает, что мы в качестве ответа выдаем разбиение выборки на классы в отношении $p:1-p$. 

Пусть настоящее отношение количества класса $0$ к $1$ равно $a : b, a+b=1$. Тогда при определенном $p$:

```
      | positive | negative |
------+----------+----------
true  |    pa    |    pb    |
------+----------+----------
false |  (1-p)a  |  (1-p)b  |
------ ---------- ----------

```


По мере роста $FPR = \frac{FP}{FP + TP} = \frac{pa}{pa + (1-p)a} = p =\frac{pb}{pb + (1-p)b} = \frac{TP}{TP + FN}=  TPR$. Т.е. $TPR$ растет пропорционально $FPR$, а значит, график ROC выглядит примерно так:

````python
plt.plot(np.linspace(0, 1, 100), np.linspace(0, 1, 100))
plt.show()
````

и $AUROC \approx \frac{1}{2}$.

** Задача 2**
Покажите, что с ростом размерности пространства признаков при равномерном распределении точек в кубе $[0; 1]^d$ вероятность попасть в куб $[0; 0.99]^d$ стремится к нулю. Это одна из иллюстраций проклятия размерностей (dimension curse). Попробуйте придумать или найти еще какую-нибудь иллюстрацию к этому явлению и кратко изложить.

**Решение:** Достаточно увеличивать $d$ и рассматривать вероятность попасть в куб.

При $d = 1$ вероятность равна $0.99$.

При росте $d$ вероятность попадания в куб убывает как $0.99^d$, а это число стремится к $0$.

Еще один пример проклятия размерности: в методе Монте-Карло для достижения удовлетворительного приближения методом Монте-Карло для интеграла от функции, принимающей большие  значения на "тощих" многомерных областях, требуется очень много псевдослучайных точек.

Например, вероятность случайного попадания в k-мерный шар, вписанный в k-мерный единичный куб, при равномерном случайном выборе точки, равна 
$$ p_k = \cfrac{\left(\sqrt{\pi} / 2\right)^k}{\Gamma\left(\cfrac{k+2}{2}\right)}$$

И это число стремительно падает при росте $k$.

#### Предобработка данных

````python
def number_encode_features(init_df):
    result = init_df.copy()
    encoders = {}
    for column in result.columns:
        if result.dtypes[column] == np.object:
            encoders[column] = preprocessing.LabelEncoder()
            result[column] = encoders[column].fit_transform(result[column])
    return result, encoders

encoded_data, encoders = number_encode_features(df)
encoded_data.sample(10)
````

Удалим пропуски в данных

````python
encoded_data.replace(' ?', np.nan).groupby('workclass').count()[['salary']].index
encoded_data.replace(' ?', np.nan, inplace=True)
encoded_data = encoded_data.dropna()

# encoded_data = encoded_data[['age', 'race', 'sex', 'marital-status', 'salary']]
````

#### Train/test split

````python
train, test = train_test_split(encoded_data, test_size=test_size)
n_test = len(test)
print('Размеры выборок: train', train.shape, '; test', test.shape)
````

````python
test.head()
````

````python
# Для того чтобы оценить bias и variance, нужно сгенерировать случайные подвыборки 
# с возвращением из исходного массива данных (бутстрепинг)

X_train = []
y_train = []
for i in range(n_repeat):
    sample_train = train.sample(len(train), replace=True) 
    X, y = sample_train.drop(['salary'], axis=1), sample_train['salary']
    X_train.append(X)
    y_train.append(y)

X_test, y_test = test.drop(['salary'], axis=1), test['salary']
````

````python
# Для каждого значения парамера оцените значение error, bias и variance согласно следующей схеме
p_name = 'p'
bias_variance_df = pd.DataFrame.from_dict({p_name:parameters[p_name]
                                           , 'bias':list([np.NaN]*len(parameters[p_name]))
                                           , 'variance':list([np.NaN]*len(parameters[p_name]))
                                           , 'error':list([np.NaN]*len(parameters[p_name]))})
````

````python
for parameter in tqdm(parameters[p_name]):
    print('parameter :', parameter)
    # Compute predictions
    y_predict = np.zeros((n_test, n_repeat))

    for i in range(n_repeat):
        estimator = KNeighborsClassifier(p=parameter, n_jobs=-1).fit(X_train[i], y_train[i])
        y_predict[:, i] = estimator.predict(X_test)

    # Bias^2 + Variance + Noise decomposition of the mean squared error
    y_error = np.sum(((np.tile(y_test, n_repeat).reshape((len(y_test), n_repeat))\
                       - y_predict) ** 2), axis=1) / n_repeat

    # В каждой точке тестовой выборки bias положим равным квадрату разности реального значения 
    # и среднего предсказанного значения по всем обучающим выборкам.
    y_bias = (y_test - np.mean(y_predict, axis=1)) ** 2
    
    # Variance положим равным среднеквадратичному разбросу значений по всем обучающим выборкам. 
    # Итоговое значение bias и variance возьмём равным среднему по всем точкам тестовой выборки
    y_var = np.var(y_predict, axis=1)
    
    #сохраните среднее значение bias^2 для данного значения параметра
    bias_variance_df.loc[bias_variance_df[p_name]==parameter, 'bias'] = y_bias.mean() 
    
    # сохраните среднее значение variance для данного значения параметра
    bias_variance_df.loc[bias_variance_df[p_name]==parameter, 'variance'] = y_var.mean()
    
    # сохраните среднее значени error для данного значения параметра
    bias_variance_df.loc[bias_variance_df[p_name]==parameter, 'error'] = y_error.mean()
    
# нарисуйте зависимость bias, variance и error от значений параметров
plt.figure(figsize=(10, 8))
bias_variance_df.set_index(p_name).plot()
````

<h2 align="center">Реализуйте kNN (2 балла)</h2>

````python
import warnings
import numpy as np
from scipy.spatial.distance import cdist

warnings.simplefilter("ignore")

class kNNClassifier():
    def __init__(self, n_estimators, metric):
        """
        Parameters
        ----------
        n_neighbours: int
            Число соседей

        metric: *alias
            метрика измерения расстояний

          """
        self.n_neighbours = n_estimators
        self.metric = metric
        
    def fit(self, X, y):
        """
        
        Parameters
        ----------
        X: 2d np.array
        y: 1d np.array
        """

        # Тут храните описание объектов обучающей выборки
        self.X_learn = X

        # Тут храните ответы по каждому объекту обучающей выборки
        self.y_learn = y
        
        return self

    def predict(self, X):
        """

        Parameters
        ----------
        X: 2d np.array матрица объекты признаки на которых нужно сказать ответ

        Returns
        -------
        y_pred: 1d np.array, вектор классов для каждого объекта
        """
        
        X = pd.DataFrame(X)
        X_len = len(X)
               
        ind1 = np.argsort(cdist(X, self.X_learn, metric=self.metric), axis=1)[:, :self.n_neighbours].ravel()
        dists = mode(self.y_learn.iloc[ind1].reshape((X_len, self.n_neighbours)), axis=1).mode.ravel()
#         print(' >>>>> dists:', dists)
#         print(dists.shape)
        
        return np.array(dists)
````

````python
def calculate_Q(estimator, X_train, y_train, X_test, y_test, feature_set):
    estimator.fit(X_train[list(feature_set)], y_train)
    return 1 - accuracy_score(estimator.predict(X_test[list(feature_set)]), y_test)
````

````python
def add_del(estimator, X_train, y_train, X_test, y_test, d):
    X_train = pd.DataFrame(X_train)
    X_test = pd.DataFrame(X_test)
    G = {}
    Q_old = None
    Q = float('+inf')
    Q_star = float('+inf')
    global_Q = None
    feature_set_star = set()
    n = X_train.shape[1]
    t, t_star = 0, 0
    
    all_features = set(X_train.columns)
    curr_feature_set = set()
    
    print('Entering cycle...')
    while True:
        
        Q_old = Q
        
        print('\nIn cycle! Q_old =', Q_old)
        
        # TRY ADD ------------------------------------------------------

        print('---------------------- ADD STAGE ----------------------')
        while len(curr_feature_set) < n:
            t += 1
            
            print('add cycle ...')

            # find f* := argmin ...: -------------------------------
            avaiable_features = all_features.difference(curr_feature_set)
#             print('..itering by :', avaiable_features)
            best_Q, best_feature = float('+inf'), None
            
            for f_iter in avaiable_features:
                
                new_poss_feature_set = curr_feature_set | set([f_iter]) # add new possible feature
                new_Q = calculate_Q(estimator, X_train, y_train, X_test, y_test, new_poss_feature_set)
#                 print('....it :', f_iter, 'newQ :', new_Q, 'with features ', new_poss_feature_set)
                if new_Q < best_Q:
                    best_Q, best_feature = new_Q, f_iter

            # looking if something has changed
            curr_feature_set = curr_feature_set | set([best_feature])
            print('after argmin feature set', curr_feature_set, 'with Q =', best_Q)
            if (best_Q < Q_star):
                t_star, Q_star, feature_set_star = t, best_Q, curr_feature_set
            if t - t_star >= d:
                break
                    
        # TRY DEL -------------------------------------------------------
        
        print('---------------------- DEL STAGE ----------------------')
        while len(curr_feature_set) > 0:
            t += 1

            print('del cycle ...')
            
            # find f* := argmin ...: -------------------------------
            avaiable_features = curr_feature_set
            
#             print('..itering by :', avaiable_features)
            
            best_Q, best_feature = float('+inf'), None
            for f_iter in avaiable_features:
                new_poss_feature_set = curr_feature_set - set([f_iter]) # remove possible feature
                new_Q = calculate_Q(estimator, X_train, y_train, X_test, y_test, new_poss_feature_set)
#                 print('....it :', f_iter, 'newQ :', new_Q, 'with features ', new_poss_feature_set)
                if new_Q < best_Q:
                    best_Q, best_feature = new_Q, f_iter

            # looking if something has changed
            curr_feature_set = curr_feature_set | set([best_feature])
            print('after argmin feature set', curr_feature_set, 'with Q =', best_Q)
            if (best_Q < Q_star):
                t_star, Q_star, feature_set_star = t, best_Q, curr_feature_set
            if t - t_star >= d:
                break
                
        #  ------------------------------------------------------------
        
        Q = Q_star
        if (Q >= Q_old):
            break
    return {'best_set' : feature_set_star,
            'dropped_set' : all_features - feature_set_star,
            'Q' : Q_star}
````
